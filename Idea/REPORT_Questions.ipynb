## What percentage of reference questions are being asked between 5 pm and 9   am in each location?

```python
def percentage(part, whole):                  
	return 100 * float(part)/float(whole)		

q = backend.QueryBuilder()                    # Create a new custom query
q.set_columns(['Question_Time', 'Question','Location'])  # we are interested specified columns
q.show = False                                # not displaying the result of this query on screen 
q.limit = 0                                   # allows us to take out the result limitations 
times = q.execute(con)

timef = "%I:%M:%S %p"
whole = 0
part = 0 
for each in times:   
    if time.strptime(each[0], timef) > time.strptime("05:00:00 PM", timef): 
        part += 1
    if time.strptime(each[0], timef) < time.strptime("09:00:00 AM", timef):
        part += 1
    whole +=1

print 'About', percentage(part, whole),'% of the questions are being asked in between 5pm and 9am'

infostats = [('between 5pm and 9am', part), ('Working hours',whole-part)]
pie_chart(infostats)
```
